```python
import pandas as pd

df = pd.read_csv('survey lung cancer.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
data = df

data["GENDER"].replace(["M", "F"], [0, 1], inplace=True)
data["LUNG_CANCER"].replace(["NO", "YES"], [0, 1], inplace=True)
data = pd.get_dummies(data)
data
```

```python
from sklearn.preprocessing import StandardScaler
predictions = ['GENDER', 'AGE', 'SMOKING', 'YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE', 'CHRONIC DISEASE', 'FATIGUE ',
               'ALLERGY ', 'WHEEZING', 'ALCOHOL CONSUMING', 'COUGHING', 'SHORTNESS OF BREATH', 'SWALLOWING DIFFICULTY', 'CHEST PAIN']

X = data[predictions]
y = data['LUNG_CANCER']

scaler = StandardScaler()
scaler.fit(X)
scaled_data = scaler.transform(X)
scaled_data = pd.DataFrame(scaled_data, columns=X.columns)
scaled_data.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    scaled_data, y, test_size=0.2, random_state=1)
print(len(y_test))
```

```python
from sklearn import linear_model

model = linear_model.LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
