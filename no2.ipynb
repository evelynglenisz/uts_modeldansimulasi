```python
import pandas as pd
from sklearn.datasets import load_iris

# Membaca file iris.csv
survey = pd.read_csv('survey lung cancer.csv')
```

```python
survey.info()

survey.head()
```

```python
survey["GENDER"].replace(["M", "F"], [0, 1], inplace=True)
survey
```

```python

from sklearn.model_selection import train_test_split
X = survey[['GENDER', 'AGE', 'SMOKING', 'YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE', 'CHRONIC DISEASE', 'FATIGUE ', 'ALLERGY ', 'WHEEZING', 'ALCOHOL CONSUMING', 'COUGHING', 'SHORTNESS OF BREATH', 'SWALLOWING DIFFICULTY', 'CHEST PAIN']]
y = survey['LUNG_CANCER']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=123)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree_model = DecisionTreeClassifier()

tree_model = tree_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

y_pred = tree_model.predict(X_test)

acc_secore = round(accuracy_score(y_pred, y_test), 3)

print('Accuracy: ', acc_secore)
print('classification report')
print(classification_report(y_pred, y_test))
```

```python
from sklearn.tree import export_graphviz
export_graphviz(
    tree_model,
    out_file="survey_tree.dot",
    feature_names=['GENDER', 'AGE', 'SMOKING', 'YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE', 'CHRONIC DISEASE', 'FATIGUE ',
                   'ALLERGY ', 'WHEEZING', 'ALCOHOL CONSUMING', 'COUGHING', 'SHORTNESS OF BREATH', 'SWALLOWING DIFFICULTY', 'CHEST PAIN'],
    class_names=['YES', 'NO'],
    rounded=True,
    filled=True
)
```
